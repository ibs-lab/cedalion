```python
import numpy as np
import pandas as pd
import xarray as xr


# Create a time range with 100 points, starting from a reference, with 0.5s steps
time = pd.date_range('2024-03-11', periods=100, freq='0.5S')

# Create an array of integers from 1 to 100
data = np.arange(1, 101)

# Combine into an xarray DataArray
xarr = xr.DataArray(data, coords=[('time', time)])

ii = 2

diff_np = abs(data[ii:] - data[:-ii])

# set last ii elements to 0 across all dimensions
xarr_zero =xarr.copy()
strt_zeroidx = xarr_zero.time[-ii]
xarr_zero.loc[dict(time=slice(strt_zeroidx, None))] = 0

xarr_shift = xarr.shift(time=-ii, fill_value=0)

# diff xr
diff_xr = abs(xarr_zero - xarr_shift)

diff_xr
```
